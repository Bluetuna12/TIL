# Mission 2. 가즈아!

서울의 소시민 나일론 마스크는 요즘 가상화폐에 푹 빠져있다  
매일매일 극락과 지옥을 오가는 매운맛에 정신을 못 차리고 있는데 그의 친구인 호주니는 그에게 정신차리게 하기 위해 비트코인 광풍이 일었던 2017년 한 해의 가격의 변화를 보여주려고 한다  
그런데 매일매일의 데이터를 보여주는 것보다 이것의 추세를 표현해주면 좋겠다는 생각이 들어 **Moving Average(이동평균법)** 를 도입하고자 한다  
호주니를 도와 마스크씨를 설득해보자

```python
bit = pd.read_csv('./dataset/BitCoin.csv')
eth = pd.read_csv('./dataset/ETH_day.csv')
```

## mandatory Part

2016.6 ~ 2017.6 기간의 **5-MA(Moving Average) 비트코인 가격 그래프**를 그려주세요  

- 선의 색깔은  **노란색(#f2a900)**으로 해야합니다
- x-axis label과 y-axis label은 적절한 텍스트로 추가되어야 합니다
- 이동평균의 기준은 `Open` Column으로 계산합니다
- 이외에도 그래프를 알아보기 쉽게 하기 위한 정보(Text, Facet 등)를 추가하는 것도 좋습니다

```python
bit_ma = bit.loc[::-1].reset_index(drop=True)
bit_ma['Date'] = pd.date_range(start='28/04/2013', end='09/22/2017')
bit_ma['5-MA'] = bit_ma['Open'].rolling(window=5).mean()
bit_ma = bit_ma[(pd.to_datetime('2016-06-01') <= bit_ma['Date']) & (pd.to_datetime('2017-06-30') >= bit_ma['Date'])]

plt.plot(bit_ma['Date'], bit_ma['5-MA'], label='5-MA', color='#f2a900')
plt.title('Bit Coin')
plt.xlabel('Date')
plt.ylabel('5-MA(Moving Average)')

plt.show()
```

## Bonus Part

2016.6 ~ 2017.6 기간에 해당하는 **비트코인의 5-MA 그래프와 이더리움의 5-MA 그래프를 동세에** 그려주세요

- 선의 색깔은 비트코인 : **노란색(#f2a900)**, 이더리움 : **검은색(#3c3c3d)** 으로 해야합니다
- x-axis label과 y-axis label은 적절한 텍스트로 추가되어야 합니다
- 이동평균의 기준은 `Open` Column으로 계산합니다
- 이외에도 그래프를 알아보기 쉽게 하기 위한 정보(Text, Facet 등)를 추가하는 것도 좋습니다

```python
eth_ma = eth.loc[::-1].reset_index(drop=True)
eth_ma['Date'] = pd.date_range(start='05/09/2016', end='04/15/2020')
eth_ma['5-MA'] = eth_ma['Open'].rolling(window=5).mean()
eth_ma = eth_ma[(pd.to_datetime('2016-06-01') <= eth_ma['Date']) & (pd.to_datetime('2017-06-30') >= eth_ma['Date'])]

plt.plot(bit_ma['Date'], bit_ma['5-MA'], label='Bit Coin', color='#f2a900')
plt.plot(eth_ma['Date'], eth_ma['5-MA'], label='Ethereum', color='#3c3c3d')
plt.title('Bit Coin, Ethereum')
plt.xlabel('Date')
plt.ylabel('5-MA(Moving Average)')
plt.legend()

plt.show()
```
